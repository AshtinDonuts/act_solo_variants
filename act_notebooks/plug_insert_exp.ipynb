## 1. Motivation

Plug insertion is a highly dexteorus task.

If we can show Plug Insert does not work when we remove some camera views, this may serve as a good testing ground for tesing Force-Torque + ACT.

**Goal** : Show that ACT performance is impacted on Plug-insert when we remove certain camera views.

## 2. Hypothesis

* Plug insert is the most challenging dexterity task
* Plug insert works decently on Base ACT with all camera views.
* Plug insert deteriorates *slightly* if we remove Right Shoulder Camera.
* Plug insert completely fails without Wrist Camera.

## 3. Method
  
### 1. Train ACT in various configuration settings on the Plug-insert dataset:
* Base ACT model
* Base ACT model without right shoulder
* ACT + Simple Torque into encoder
* ACT + 6-ax Force-Torque into encoder

### 2. Run inference test for each configuration
* Height generalization
* Positional generalization
